```python
aqimedavg = sum(aqimed2020)/len(aqimed2020)
tdaysavg = sum(totaldays2020)/len(totaldays2020)
pdaysavg = sum(poordays2020)/len(poordays2020)
```

```python
totaldays2010 = [365,365,365,365,365,365,365,365,351,365,365,365,365,365,365,365,365,365,365,365]
poordays2010 = [3,14,115,1,5,38,24,59,1,5,43,103,28,23,72,10,50,6,33,8]
aqimed2010 = [41,48,77,38,32,61,51,74,24,60,71,74,52,51,63,47,55,48,44,38]
zone2010 = ['west','west','west','west','west','mountain','mountain','mountain','mountain','mountain','central','central','central','central','central','east','east','east','east','east']
```

```python
aqicities2010 = pd.DataFrame({'Total_Days2010':totaldays2010,'Poor_Days2010':poordays2010,'AQI_Median2010':aqimed2010,'Zone2010':zone2010})
```

```python
aqicities2010
```

```python
alldata = pd.concat([aqicities2020,aqicities2010], axis = 1, join = 'inner')
```

```python
alldata
```

```python
alldata['Scale2020'] = alldata['Total_Days2020']/366
```

```python
alldata
```

```python
alldata['Poorday_Rate2020'] = alldata['Poor_Days2020']/alldata['Scale2020']
```

```python
alldata['Poorday_Rate2020'] = alldata['Poorday_Rate2020'].astype(int)
```

```python
alldata
```

```python
alldata['Scale2010'] = alldata['Total_Days2010']/365
```

```python
alldata['Poorday_Rate2010'] = alldata['Poor_Days2010']/alldata['Scale2010']
```

```python
alldata['Poorday_Rate2010'] = alldata['Poorday_Rate2010'].astype(int)
```

```python
alldata.to_csv('Finalproj/alldata.csv')
```
