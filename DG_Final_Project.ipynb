```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import geopandas as gp
```

```python
midterm = pd.read_csv('Finalproj/alldata.csv')
```

```python
midterm
```

```python
midterm['lon'] = [-122.19,-122.25,-118.14,-157.51,-149.53,-104.59,-111.53,-112.04,-114.28,-106.39,-87.37,-90.11,-96.47,-82.59,-90.04,-71.03,-77.02,-80.11,-74.45,-70.15]
midterm['lat'] = [47.36,37.46,34.04,21.18,61.12,39.44,40.46,33.28,42.33,35.05,41.53,38.37,32.47,39.58,29.57,42.21,38.54,25.45,40.13,43.39]
```

```python
midterm
```

```python
cities = gp.GeoDataFrame(midterm, geometry = gp.points_from_xy(midterm.lon, midterm.lat))
cities.geometry
```

```python
filepath = 'finalproj/states.shp'
states = gp.read_file(filepath)
```

```python
states.crs
```

```python
states.head()
```

```python
states = states.to_crs("epsg:4326") #Mercator-projection
```

```python
states.crs
```

```python
cities
```

```python
fig, ax = plt.subplots(figsize=(20,20))
states.plot(ax=ax, color='lightgrey', edgecolor='black', linewidth=0.4, legend=True)
cities.plot(column='Zone2020',ax=ax, markersize='Poor_Days2020', alpha=1.0, categorical=False, legend=True)
ax.axis()
plt.title('Air Quality Index Poor Days: 2020', size=20)
plt.savefig('Finalproj/poordays2020.png')
#plt.axis()
#plt.show()
```

```python
fig, ax = plt.subplots(figsize=(20,20))
states.plot(ax=ax, color='lightgrey', edgecolor='black', linewidth=0.4, legend=True)
cities.plot(column='Zone2020', ax=ax, markersize='Poor_Days2010', alpha=1.0, categorical=False, legend=True)
ax.axis()
plt.title('Air Quality Index Poor Days: 2010', size=20)
plt.savefig('Finalproj/poordays2010.png')
#plt.axis()
#plt.show()
```

```python
import folium

#Make an empty map

m = folium.Map(location=[34,-100], tiles="OpenStreetMap", zoom_start=4)

#show the map

m
```

```python
import folium

for i in range(0,len(cities)):
   folium.Circle(
    location=[cities.iloc[i]['lat'],cities.iloc[i]['lon']],
    popup=cities.iloc[i]['City'],
    radius=float(cities.iloc[i]['Poor_Days2020'])*2000,
    color='crimson',
    fill=True,
    fill_color='crimson'
    ).add_to(m)
    
#show the map again

m
```

```python
m.save('Finalproj/Poordays2020.html')
```
